```python
import pandas as pd

book_rating_count = {}

with open('../data/goodreads_interactions.csv', 'r') as f:
    while True:
        line = f.readline()
        if not line:
            break
        _, csv_id, _, rating, _ = line.strip().split(',')
        if csv_id not in book_rating_count:
            book_rating_count[csv_id] = 1
        else:
            book_rating_count[csv_id] += 1
```

```python
len(book_rating_count)
```

```python
threshold = 500
popular_books = {book_id for book_id, count in book_rating_count.items() if count >= threshold}
```

```python
len(popular_books)
```

```python
import csv

input_file = "../data/goodreads_interactions.csv"
output_file = "filtered_interactions.csv"

with open(input_file, "r", encoding="utf-8") as infile, \
     open(output_file, "w", newline="", encoding="utf-8") as outfile:

    reader = csv.DictReader(infile)
    writer = csv.DictWriter(outfile, fieldnames=reader.fieldnames)
    writer.writeheader()

    for row in reader:
        if row["book_id"] in popular_books:
            writer.writerow(row)
```
